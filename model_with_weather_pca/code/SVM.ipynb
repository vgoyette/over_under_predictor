```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
df = pd.read_csv('../datasets/pca_weather_scores_w_avgs.csv')
df
```

```python
cols = ['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6', 'pc7', 'pc8']
features = df[cols]
features
```

```python
labels = np.ravel(df['encoded_label'])
labels
```

```python
from sklearn.model_selection import cross_validate
svc = SVC(kernel='linear', C=0.7)


cv_results = cross_validate(svc, features, labels, cv=10, scoring=('accuracy', 'precision', 'recall', 'f1'))

del cv_results['score_time']
del cv_results['fit_time']

for metric_name, values in cv_results.items():
    print(f'Metric: {metric_name}')
    #for val in values:
    #    print(f'\t{val}')
    print(f'\tMean: {np.mean(values)}')
    print(f'\tStandard Deviation: {np.std(values)}')
```

```python
svc = SVC(kernel='poly', C=0.4)


cv_results = cross_validate(svc, features, labels, cv=10, scoring=('accuracy', 'precision', 'recall', 'f1'))

del cv_results['score_time']
del cv_results['fit_time']

for metric_name, values in cv_results.items():
    print(f'Metric: {metric_name}')
    #for val in values:
    #    print(f'\t{val}')
    print(f'\tMean: {np.mean(values)}')
    print(f'\tStandard Deviation: {np.std(values)}')
```
