```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('../datasets/non_pca_weather_scores_w_avgs.csv')
df
```

```python
cols = ['weather_temperature', 'weather_wind_mph', 'home_avg_first_downs', 'home_avg_yds', 'home_avg_pass_yds', 'home_avg_rush_yds', 'home_avg_TO', 'away_avg_first_downs', 'away_avg_yds', 'away_avg_pass_yds', 'away_avg_rush_yds', 'away_avg_TO']
features = df[cols]
features
features = StandardScaler().fit_transform(features)
features
```

```python
labels = np.ravel(df['encoded_label'])
labels
```

```python
from sklearn.model_selection import cross_validate
svc = SVC(kernel='linear', C=0.7)


cv_results = cross_validate(svc, features, labels, cv=10, scoring=('accuracy', 'precision', 'recall', 'f1'))

del cv_results['score_time']
del cv_results['fit_time']

```

```python
for metric_name, values in cv_results.items():
    print(f'Metric: {metric_name}')
    #for val in values:
    #    print(f'\t{val}')
    print(f'\tMean: {100 * np.mean(values)}')
    print(f'\tStandard Deviation: {100 * np.std(values)}')
```

```python
svc = SVC(kernel='poly', C=0.4)


cv_results = cross_validate(svc, features, labels, cv=10, scoring=('accuracy', 'precision', 'recall', 'f1'))

del cv_results['score_time']
del cv_results['fit_time']
```

```python
for metric_name, values in cv_results.items():
    print(f'Metric: {metric_name}')
    #for val in values:
    #    print(f'\t{val}')
    print(f'\tMean: {100 * np.mean(values)}')
    print(f'\tStandard Deviation: {100 * np.std(values)}')
```
