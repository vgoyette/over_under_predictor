```python
import pandas as pd
df = pd.read_csv('../datasets/scores_w_avgs.csv')
df = df.dropna()
df
```

```python
df['score_home'] = df['score_home'].astype('float64')
df['score_away'] = df['score_away'].astype('float64')
df['total_score'] = df['total_score'].astype('float64')
df.info()
```

```python
features = pd.DataFrame(df, columns = ['home_avg_yards', 'home_avg_TO', 'away_avg_yards', 'away_avg_TO'])
labels = pd.DataFrame(df, columns=['label'])
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
model = LogisticRegression()

features_train, features_test, labels_train, labels_test = train_test_split(features, np.ravel(labels), test_size=0.3, random_state=109)
model.fit(features_train, labels_train)

predictions = model.predict(features_test)
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(labels_test, predictions).ravel()
print(f'True Positives: {tp}')
print(f'True Negatives: {tn}')
print(f'False Positives: {fp}')
print(f'False Negatives: {fn}')
```

```python
accuracy = (tp + tn) / (tp + fp + fn + tn)
precision = tp / (tp + fp)
recall = tp / (tp + fn) 
f1 = (2*(recall * precision)) / (recall + precision)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```
